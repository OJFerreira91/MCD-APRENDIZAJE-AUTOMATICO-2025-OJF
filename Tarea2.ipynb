```python
import pandas as pd
# cargar datos
df = pd.read_excel("estudiantes.xlsx")
df
```

```python
# Tipos de datos del set
print(df.dtypes)
```

```python
df = df.astype({
    "Nombre": "string",
    "Edad": "int",
    "Carrera": "string",
    "Promedio": "float",
    "Semestre": "int"
})

```

```python
# Descripción de los datos
df.describe()
```

```python
# Columna como índice
df = df.set_index("Nombre")
print(df.head())
```

```python
# Crear columna de estado (aprobado si promedio ≥ 8):
df["Estado"] = df["Promedio"].apply(lambda x: "Aprobado" if x >= 8 else "Reprobado")
df
```

```python
# Modificar columna (por ejemplo, aumentar en 1 todos los semestres):
df["Semestre"] = df["Semestre"] + 1
df
```

```python
# Eliminar columna:
df = df.drop(columns=["Edad"])
df
```

```python
# Estudiantes de Ingeniería con promedio mayor a 6.
filtro = df[(df["Carrera"] == "Ingeniería") & (df["Promedio"] > 6)]
filtro
```

```python
import pandas as pd
# Crear subconjuntos de datos 
# Solo nombre y promedio.
df = pd.read_excel("estudiantes.xlsx")
sub_df = df[["Nombre", "Promedio"]]
sub_df
```

```python
# Crear grupos de datos
# Agrupar por carrera y calcular promedio general:
grupos = df.groupby("Carrera")["Promedio"].mean()
print(grupos)
```
